```python
squares = []
for i in range(10):
    squares.append(i**2)

print (squares)
```

```python
squares = [i**2 for i in range(10)]
print (squares)
```

```python
even_squares = [i**2 for i in range (10) if i % 2 == 0]
print (even_squares)
```

```python
def add_all(*args):
    return sum(args)

print(add_all(1, 2, 3, 4, 5))
```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Vansh", role="AI Engineer", city="India")
```
